# Select Dataset
* Concrete Compressive Strength: https://archive.ics.uci.edu/dataset/165/concrete+compressive+strength
* Concrete is the most important material in civil engineering. The concrete compressive strength is a highly nonlinear function of age and ingredients.

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
concrete_compressive_strength = fetch_ucirepo(id=165) 
  
# data (as pandas dataframes) 
X = concrete_compressive_strength.data.features 
y = concrete_compressive_strength.data.targets 
  
# metadata 
print(concrete_compressive_strength.metadata) 
  
# variable information 
print(concrete_compressive_strength.variables) 
```

# Create Correlation Plot

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import t

# Concatenate features and target into a DataFrame
df = pd.concat([X, y], axis=1)
print(df)
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plotting the correlation matrix using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

On interpreting the correlation matrix: 
* Positive correlation is indicated by warmer colors (closer to 1). 
* Negative correlation is indicated by cooler colors (closer to -1). 
* A value closer to 0 indicates little to no correlation.

# Check the Distribution of Each Column 
* **Kernel Density Estimation (KDE)** is performed alongside histograms to provide a smoother estimate of the probability density function (PDF) of the data. While histograms give a discrete representation of the distribution by binning the data, KDE provides a continuous estimate of the underlying probability density.
*  **Kolmogorov-Smirnov (KS) test** serves the purpose of objectively evaluating the goodness of fit of different theoretical distributions to the observed data, aiding in the selection of the most appropriate distribution for describing the data's underlying probability distribution.

```python
import scipy.stats as stats

# Loop through each column in the DataFrame
for column in df.columns:
    # Plot histogram and KDE for the column
    sns.histplot(df[column], kde=True, stat='density', bins=20)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
    
    # Fit different distributions and calculate their goodness of fit
    distributions = ['norm', 'expon', 'gamma', 'lognorm']
    best_fit_name = None # normal, exponential, gamma, log-normal
    best_fit_params = None # three parameters: shape (α), scale (β), and location (θ)
    best_kstest_stat = np.inf # close to 0 indicates a good fit between the empirical data and the theoretical distribution
    
    for dist_name in distributions:
        # Fit the distribution to the data
        dist = getattr(stats, dist_name)
        params = dist.fit(df[column])
        
        # Perform Kolmogorov-Smirnov test
        kstest_stat, _ = stats.kstest(df[column], dist_name, args=params)
        
        # Update best fit if current distribution has lower KS test statistic
        if kstest_stat < best_kstest_stat:
            best_kstest_stat = kstest_stat
            best_fit_name = dist_name
            best_fit_params = params
            
    # Print the best fit distribution and its parameters
    print(f"Best fit for {column}: {best_fit_name} with parameters {best_fit_params} (KS test stat: {best_kstest_stat})")
```

# Perform a Hypothesis Test

```python
import numpy as np
from scipy.stats import t

# Define the null and alternative hypotheses
# Null Hypothesis (H0): Mean compressive strength of concrete is equal to a certain value (e.g., 50)
# Alternative Hypothesis (H1): Mean compressive strength of concrete is not equal to the certain value

mu = 50  # Certain value
alpha = 0.05  # Significance level

# Sample data (replace this with your actual data)
concrete_strength = df['Concrete compressive strength'].values

# Calculate sample mean and standard deviation
sample_mean = np.mean(concrete_strength)
sample_std = np.std(concrete_strength, ddof=1)  # Use ddof=1 for sample standard deviation

# Calculate t-statistic
n = len(concrete_strength)
t_statistic = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Calculate degrees of freedom
degrees_of_freedom = n - 1

# Calculate p-value
p_value = 2 * (1 - t.cdf(np.abs(t_statistic), df=degrees_of_freedom))  # 2-tailed test

# Print results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("t-statistic:", t_statistic)
print("Degrees of Freedom:", degrees_of_freedom)
print("p-value:", p_value)

# Decision based on p-value and significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is significant evidence to suggest that the mean compressive strength of concrete differs from", mu)
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the mean compressive strength of concrete differs from", mu)
```
