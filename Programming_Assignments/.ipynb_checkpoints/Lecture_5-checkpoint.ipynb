# Select Dataset

* Forest Fires: https://archive.ics.uci.edu/dataset/162/forest+fires
* This is a difficult regression task, where the aim is to predict the burned area of forest fires, in the northeast region of Portugal, by using meteorological and other data (see details at: http://www.dsi.uminho.pt/~pcortez/forestfires).

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
forest_fires = fetch_ucirepo(id=162) 
  
# data (as pandas dataframes) 
X = forest_fires.data.features 
y = forest_fires.data.targets 
  
# metadata 
print(forest_fires.metadata) 
  
# variable information 
print(forest_fires.variables) 
```

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Concatenate features (X) and target (y) into a DataFrame
df = pd.concat([X, y], axis=1)
print(df)
```

# Perform Regression Analysis

```python
# Apply dummy coding to categorical variables
dummy_df = pd.get_dummies(df, columns=['month', 'day'])

print(dummy_df)
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(dummy_df.drop(columns=['area']))  # Exclude the target variable
scaled_df = pd.DataFrame(scaled_features, columns=dummy_df.drop(columns=['area']).columns)

# Concatenate scaled features and target
final_df = pd.concat([scaled_df, dummy_df['area']], axis=1)

print(final_df)
```

## Visulization
* Visualize the relationship and assess whether a linear or polynomial model is more appropriate.
* **Scatter plots** to visualize the raw data and assess the relationship between variables.
* **Fitted curves** to visualize how well the model fits the data and captures the underlying pattern.

```python
# Select columns of interest
selected_columns = ['X', 'Y', 'FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']
```

```python
# Scatter plots
sns.pairplot(final_df, x_vars=selected_columns, y_vars=['area'], kind='scatter')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Fitted curves - Simple Linear Regression
for feature in selected_columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=scaled_df[feature], y=final_df['area'], color='blue', alpha=0.5)
    sns.regplot(x=scaled_df[feature], y=final_df['area'], scatter=False, color='red', order=1, ci=None, label='Linear')
    plt.xlabel(feature)
    plt.ylabel('Area')
    plt.legend()
    plt.title(f'Fitted Curve - {feature} vs. Area (Linear)')
    plt.show()

# Fitted curves - Polynomial Regression (e.g., quadratic)
for feature in selected_columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=scaled_df[feature], y=final_df['area'], color='blue', alpha=0.5)
    
    # Polynomial regression
    polynomial_features = PolynomialFeatures(degree=2)
    X_poly = polynomial_features.fit_transform(scaled_df[[feature]])
    poly_model = LinearRegression()
    poly_model.fit(X_poly, final_df['area'])
    y_poly_pred = poly_model.predict(X_poly)
    
    # Sort the values for plotting
    sorted_zip = sorted(zip(scaled_df[feature], y_poly_pred))
    x_sorted, y_poly_pred_sorted = zip(*sorted_zip)
    
    plt.plot(x_sorted, y_poly_pred_sorted, color='red', label='Quadratic')
    plt.xlabel(feature)
    plt.ylabel('Area')
    plt.legend()
    plt.title(f'Fitted Curve - {feature} vs. Area (Quadratic)')
    plt.show()
```
